# JSON & APIs

````python
import requests
from pprint import pprint
````

````python
API = "https://restcountries.com/v3.1/all"
response = requests.get(API, verify=False)

countries = response.json()

countries
````

# Task 1

Print all the country names in a list

````python
def get_country_names(countries):
    return [country['name']['common'] for country in countries]

pprint(get_country_names(countries))
````

# Task 1.2
Population more than 10_000_000 & Sorted by population

````python
def get_populated_country_names(countries, population):
    country_populations = sorted(
        {
            country["name"]["common"]: country["population"]
            for country in countries
            if country["population"] >= population
        }.items(),
        key=lambda x: x[1],
        reverse=True,
    )

    return [country_info[0] for country_info in country_populations]

print(get_populated_country_names(countries, 10_000_000))
````

3. Russia -  80,000,000
4. ...

````python
def get_formatted_countries(countries, population):
    country_populations = sorted(
        {
            country["name"]["common"]: country["population"]
            for country in countries
            if country["population"] >= population
        }.items(),
        key=lambda x: x[1],
        reverse=True,
    )

    return "\n".join(
        [
            f"{i+1:>2}. {country_info[0]:<21} {'-':<10} {country_info[1]:>20,}"
            for i, country_info in enumerate(country_populations)
        ]
    )


print(get_formatted_countries(countries, 10_000_000))
````

````python
def get_formatted_countries(countries, population, limit):
    country_populations = sorted(
        {
            country["name"]["common"]: country["population"]
            for country in countries
            if country["population"] >= population
        }.items(),
        key=lambda x: x[1],
        reverse=True,
    )

    return "\n".join(
        [
            f"{i + 1:>2}. {country_info[0]:<21} {'-'} {country_info[1]:>20,}"
            for i, country_info in enumerate(country_populations) if i < limit
        ]
    )


print(get_formatted_countries(countries, 10_000_000, 10))
````
