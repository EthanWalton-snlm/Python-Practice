# Visualize

- Matplot
  - Pioneer
  - Beginner friendly
- Seaborn
  - Uses matplot internally
  - Complex visualization easily

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

tips_df = pd.read_csv("../Day 20/tips.csv")
```

### Bar Graph

```python
y_axis = tips_df['tip'].head(5)
x_axis = np.arange(1, 6)

plt.bar(x_axis, y_axis, color="plum", alpha=0.8)
plt.xlabel('Days')
plt.ylabel('Tips')

plt.title('Tips on first five days')
```

### Task

Draw a bar graph for Average Total Bill each day

```python
y_axis = tips_df.groupby("day")["total_bill"].mean()
x_axis = y_axis.index

plt.bar(x_axis, y_axis, color="green", alpha=0.5)
plt.xlabel("Days")
plt.ylabel("Average Total Bill")
plt.title("Average Total Bill Each Day")
```

```python
y_avg_total_bill = tips_df.groupby("day")["total_bill"].mean()

y_avg_total_bill.plot(
    kind="line",
    xlabel="Days",
    ylabel="Average Total Bill",
    title="Average Total Bill Each Day",
    color="green",
    alpha=0.5,
)
```

### Task: Proportion of Meals
Create a pie chart comparing how many times people visited on which time of the day (Dinner, Lunch)

```python
y_time_visited = tips_df.groupby("time")["total_bill"].count()

y_time_visited.plot(
    kind="pie", title="Lunch vs Dinner Visitors", ylabel="", autopct='%1.1f%%'
)
```

```python
plt.scatter(tips_df["total_bill"], tips_df["tip_percent"], color="plum")

plt.xlabel("Total Bill  (R)")
plt.ylabel("Tip percent %")
plt.title("Tip percent vs Total Bill")
```

```python
y_avg_total_bill = tips_df.groupby("total_bill")["total_bill"].mean()

y_avg_total_bill.plot(
    kind="hist",
    xlabel="Total Bill (R)",
    ylabel="Freqeuncy",
    title="Average Total Bill Each Day",
    color="green",
    alpha=0.5,
    bins=15
)
```

# Seaborn

```python
y_axis = tips_df["tip"].head(4)
x_axis = np.arange(1, 5)

sns.barplot(x="day", y="total_bill", data=tips_df)
plt.xlabel("Days")
plt.ylabel("Average Total Bill")
plt.title("Average Total Bill Each Day")
plt.xticks(rotation=45)
```

```python
y_axis = tips_df["tip"].head(4)
x_axis = np.arange(1, 5)

sns.regplot(x="total_bill", y="tip", data=tips_df)
plt.xlabel("Total Bill (R")
plt.ylabel("Tip (R)")
plt.title("Total Bill vs Tip with Regression Line")
plt.xticks(rotation=45)
```
