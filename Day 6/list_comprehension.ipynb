# List Comprehension

## Pythonic Code



```python
nums = [3, 5, 10, 12]

squares = []

for num in nums:
    squares.append(num**2)

squares
```

```python
nums = [3, 5, 10, 12]

squares = [num**2 for num in nums]

squares
```

```python
odds = [num for num in range(50) if num % 2 != 0]

print(odds)
```

```python
avengers = [
    "Hulk",
    "Iron man",
    "Black widow",
    "Captain america",
    "Spider man",
    "Thor",
]

avengers_letter_count = [len(avenger) for avenger in avengers]

print(avengers_letter_count)
```

```python
points = [(3, 4), (6, 12), (10, 13)]

distances = [round((x**2 + y**2) ** 0.5, 2) for x, y in points]

print(distances)
```

```python
# Task - With List Comp

nums = [7, 44, 90, 6, 75, 10]

output = ["Odd" if num % 2 == 1 else "Even" for num in nums]

output
```
