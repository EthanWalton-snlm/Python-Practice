## Pre-Processing
1. Features Selection
2. Cleaning
3. Standardize

```python
features_df = df[["Pclass", "Age", "Fare"]]
target_df = df["Survived"]

features_df
```

```python
features_df.isnull().sum()
```

```python
features_df.describe()
```

```python
imputer = SimpleImputer(strategy="mean")
columns = ['Pclass', 'Age', 'Fare']
features_df[columns] = imputer.fit_transform(features_df[columns])

features_df
```

```python
features_df.isnull().sum()
```

```python
# 3. Standardize

from sklearn.preprocessing import StandardScaler # (3, 3) -> mean - 0, std - 1
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_df)
features_scaled # minmax - (0, 1)
```

# 4. Feature Engineering
- Bucketing (skip)
- Encoding (string) -> number

```python
target_df
```

```python
# 5. Modelling

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features_scaled, target_df, test_size=0.2, random_state=42)
```

```python
print(len(X_train), len(X_test), len(y_train), len(y_test), len(X_train) + len(X_test))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_test)

predict
```

```python
y_test # actual
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predict)

print(f"Model accuracy is: {accuracy:.2%}")
```
